### **Lesson 1: Orientation to Linear Algebra**
This lesson serves as an introduction to why linear algebra is the mathematical heart of machine learning and deep learning.

*   **Key Subtopics:**
    *   Defining Linear Algebra and its difference from general algebra.
    *   Solving systems of linear equations algebraically and graphically.
    *   Historical origins and contemporary ML applications.
*   **Core Concepts:**
    *   **Definition:** Algebra involves arithmetic with non-numerical entities (like $x$); linear algebra specifically excludes **exponential terms** (like $x^2$ or $\sqrt{x}$) to describe only straight lines on a graph.
    *   **Systems of Equations:** Linear algebra is the ubiquitous approach for solving for multiple unknowns within high-dimensional spaces simultaneously.
    *   **Possible Solutions:** In a linear system, there are only three possibilities: **one solution** (lines intersect at a point), **no solution** (parallel lines), or **infinite solutions** (identical lines).
*   **Applications:**
    *   **Regression:** Used to predict continuous values, such as house prices, by solving for unknown parameters ($a, b, c \dots m$).
    *   **Deep Learning:** Tensors store the **weights and biases** that mapping pixel inputs to categorical predictions.
    *   **Other uses:** Ranking results (Google PageRank), recommender systems (Netflix), and Natural Language Processing (NLP).

### **Lesson 2: Data Structures for Algebra**
This lesson details the fundamental building blocks of linear algebra, known as tensors.

*   **Key Subtopics:**
    *   Tensors of various Ranks (Scalars, Vectors, Matrices).
    *   Vector Transposition and Norms (L1, L2, Max Norm).
    *   Basis, Orthogonal, and Orthonormal Vectors.
*   **Core Concepts:**
    *   **Tensors:** A machine learning generalization of arrays to any number of dimensions.
        *   **Scalar (Rank 0):** A single number representing magnitude only.
        *   **Vector (Rank 1):** An ordered list of numbers representing magnitude and direction.
        *   **Matrix (Rank 2):** A 2D array of numbers (rows $\times$ columns).
    *   **Vector Transposition:** Using the notation $x^T$ to flip a row vector into a column vector (or vice versa) while keeping elements in the same order.
    *   **Norms (Magnitude):**
        *   **$L^2$ Norm (Euclidean):** The most common norm, measuring direct distance from the origin: $\|x\|_2 = \sqrt{\sum |x_i|^2}$.
        *   **$L^1$ Norm (Manhattan):** The sum of absolute values, useful for distinguishing between zero and non-zero values.
        *   **Max Norm ($L^\infty$):** Returns the absolute value of the largest element in the vector.
    *   **Orthogonality:** Two vectors are orthogonal if they are at a **90-degree angle** and their dot product is zero ($x^T y = 0$). **Orthonormal** vectors are orthogonal and have a unit norm of 1.

### **Lesson 3: Common Tensor Operations**
This lesson transitions from static data structures to the mathematical manipulations performed on tensors in ML models.

*   **Key Subtopics:**
    *   Tensor Transposition and Basic Arithmetic.
    *   The Hadamard Product vs. the Dot Product.
    *   Reduction operations.
*   **Core Concepts:**
    *   **Matrix Transposition:** Flipping axes over the **main diagonal** (top-left to bottom-right); element $i,j$ becomes $j,i$.
    *   **Basic Arithmetic:** Adding or multiplying a tensor by a **scalar** applies the operation to every element individually.
    *   **Hadamard Product ($A \odot X$):** Also known as the element-wise product; two tensors of the **same size** are multiplied element-by-element.
    *   **The Dot Product ($x \cdot y$):** Multiplies corresponding elements of two vectors and sums them into a single **scalar value** ($x \cdot y = \sum_{i=1}^{n} x_i y_i$). This is performed at every artificial neuron in a deep neural network.
    *   **Reduction:** Aggregating elements of a tensor to a lower dimension (e.g., summing all elements to a scalar or summing along a specific axis).

### **Lesson 4: Solving Linear Systems**
This lesson focuses on manual algebraic techniques to solve for unknowns before moving to computational matrix methods.

*   **Key Subtopics:**
    *   The Substitution Strategy.
    *   The Elimination Strategy.
    *   Geometric visualization of systems.
*   **Core Concepts:**
    *   **Substitution:** Useful when one variable has a coefficient of one; isolate that variable and substitute its value into the other equation.
    *   **Elimination:** Manipulate equations to eliminate one variable by adding or subtracting the equations together, then solve for the remaining unknown.
    *   **Matrix Representation:** Systems of equations can be written compactly in the form **$Ax = b$**, where $A$ is the coefficient matrix, $x$ is the vector of unknowns, and $b$ is the output vector.
    *   **Geometric Insight:** Solving for unknowns is the geometric equivalent of finding the **point of intersection** where lines (or planes in higher dimensions) cross over.
